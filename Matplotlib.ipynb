MAKING THE PRICE,AGE LIST AND MAKING SCATTER PLOT AND MARKING WITH RED COLOR AND GIVING THE TITLE AGE V/S PRICE  

```python
price=list(df["price"].values)
age=list(df["age"].values)
plt.scatter(age,price,color="r",marker="*") #marker p,d,v
plt.xlabel("AGE")
plt.ylabel("PRICE")
plt.title("AGE V/S PRICE")
plt.show()
```

FILTERING THE ROYAL ENFIELD DATA WITH PRICE LESS THEN 50000 AND SORT BY PRICE AND AGE AND PLOT THE LINEPLOT

```python
royal = df[(df["brand"]=="Royal Enfield")&(df["price"]<=50000)]
price = list(royal["price"].sort_values())
age = list(royal["age"].sort_values())

pt = plt.plot(price,age,marker ="." ,ms=16)
plt.xlabel("PRICE")
plt.ylabel("AGE")
```

BARGRAPH

MAKING THE BRAND LIST AND NO_OF_COUNT AND PLOT THE BARGRAPH WITH FIGSIZE(WIDTH,HEIGTH) AND GIVE X_LABEL AS BRAND, Y_LABEL AS NO_OF_COUNT AND GIVING TITLE AS NO_OF_BIKE_IN_BRANDS AND ROTATION THE X_LABEL AS 90 

```python
brand=list(df["brand"].value_counts().keys()) 
no_of_counts=list(df["brand"].value_counts().values)
plt.figure(figsize=(16,8))   #(width,heigth)
plt.bar(brand,no_of_counts)
plt.xlabel("BRAND")
plt.ylabel("NO_OF_COUNTS")
plt.title("NO_OF_BIKE_IN_BRANDS")
plt.xticks(rotation=90)
plt.show()
# bargraph
```

COUNTING THE BRAND 

```python
df["brand"].value_counts()
```

MAKING THE BRAND LIST AND NO_OF_COUNT AND PLOT THE BARGRAPH WITH FIGSIZE(WIDTH,HEIGTH) AND GIVE X_LABEL AS BRAND, Y_LABEL AS NO_OF_COUNT AND GIVING TITLE AS NO_OF_BIKE_IN_BRANDS AND ROTATION THE X_LABEL AS 90  GIVING SOME VISUAL(AGE LINE,EDGE COLOR AND LINE WIDTH)

```python
brand=list(df["brand"].value_counts().keys()) 
no_of_counts=list(df["brand"].value_counts().values)
plt.figure(figsize=(16,8))   #(width,heigth)
plt.bar(brand,no_of_counts,color="green",ec="r",lw=5,ls="solid",width=0.9)   #age line ,ec=age color,lw=line width, ls=dashed dotted solid
plt.xlabel("BRAND")
plt.ylabel("NO_OF_COUNTS")
plt.title("NO_OF_BIKE_IN_BRANDS")
plt.xticks(rotation=90)
plt.show()
```
